# Question 2: Maximum likelihood estimate (10 pts)

Consider the following univariate probability distribution 
$$p(x|a,b) = \frac{x^{a-1}}{\Gamma(a)b^a}exp(-\frac{x}{b}),$$ for $x \geq 0$ and $a,b > 0.  ~~ \Gamma(a)$ is a normalization term. 

You are given N i.i.d. data points drawn from this distribution. Derive the MLE estimate for $b$, in terms of $x_1, x_2, \ldots x_N$ and $a$.

## Answer



It follows,
$$ln(L(x|a,b))= ln(\frac{\prod_{i=1}^{N} x_{i}^{a-1}}{\Gamma(a)^Nb^{Na}}exp(\sum_{i=1}^{N} -\frac{x}{b})) = ln(\frac{\prod_{i=1}^{N} x_{i}^{a-1}}{\Gamma(a)^Nb^{Na}}) + ln(exp(\sum_{i=1}^{N} -\frac{x_{i}}{b})) =$$

$$= ln(\prod_{i=1}^{N} x_{i}^{a-1}) - ln(\Gamma(a)^Nb^{Na}) + \sum_{i=1}^{N} -\frac{x_{i}}{b} =$$

$$= \sum_{i=1}^{N} ln(x_{i}^{a-1}) - ln(\Gamma(a)^Nb^{Na}) + \sum_{i=1}^{N} -\frac{x_{i}}{b} =$$

$$= - ln(\Gamma(a)^Nb^{Na}) + \sum_{i=1}^{N}( -\frac{x_{i}}{b} + ln(x_{i}^{a-1}) )$$

To maximize, one first takes the derivative,

$$\frac{\partial ln(L(x|a,b))}{\partial b} = \frac{\partial }{\partial b} - ln(\Gamma(a)^Nb^{Na}) + \frac{\partial }{\partial b} \sum_{i=1}^{N}( -\frac{x_{i}}{b} + ln(x_{i}^{a-1}) ) = $$

$$ = -\frac{Na}{b} + \sum_{i=1}^{N}\frac{x_{i}}{b^2} $$

Next, the derivative is set to zero.

So if,

$$\frac{\partial ln(L(x|a,b))}{\partial b} = 0 $$ 

it follows

$$ -\frac{Na}{b} + \sum_{i=1}^{N}\frac{x_{i}}{b^2} = 0 $$

Then, one can solve for b.

$$ \frac{Na}{b} = \sum_{i=1}^{N}\frac{x_{i}}{b^2} $$

$$ Na = \sum_{i=1}^{N}\frac{x_{i}}{b} $$

$$ b = \sum_{i=1}^{N}\frac{x_{i}}{Na} $$

```python
##first Let's find the in-sample MSE and then we'll find the out of sample MSE and RMSE
ypred = results.predict(X)
MSE_insample = mean_squared_error(y, ypred)
print 'Coefficients:'
print results.params
print "The MSE on the training data is:", MSE_insample

#Out of Sample Prediction
y, X = dmatrices('Sales ~ TV + Radio + Newspaper', data=test, return_type='dataframe')
ypred = results.predict(X)
MSE_OOS = mean_squared_error(y, ypred)
RMSE = sqrt(MSE_OOS)
print "The MSE on the test data is:", MSE_OOS
print "The MSE goes up by", MSE_OOS - MSE_insample, "when you go from train to test data"
print "The RMSE is:", RMSE
```

# Question 6: Illustrating the curse of dimensionality (10 pts)

For a hypersphere of radius $a$ in $d$ dimensions, the volume is related to the surface area of a unit hypersphere (S) as
    $$ V = \frac{S \times a^d}{d}.$$


1. (6 pts) Use this result to show that the fraction of the volume which lies at values of the radius between $a - \epsilon$ and $a$, where $0 < \epsilon < a$, is given by
$f = 1 - (1 - \epsilon/a)^d$. Hence, show that for any fixed $\epsilon$, no matter how small, this fraction tends to 1 as $d \rightarrow \infty$.
2. (4 pts) Evaluate the ratio $f$ numerically, with $\epsilon/a = 0.01$, for the cases $d$ = 2, 10, and 1000.


This problem shows that if data are randomly distributed within a very high-dimensional hypersphere, most of the mass is near the surface and there is very little of it near the "core". Thus, there are few "representative" points.

## Answer



## Part A

## 1

##### Part a:
$$ f = \frac{V(a- \epsilon)}{V(a)} = \frac{\frac{S \times (a- \epsilon)^d}{d}}{\frac{S \times a^d}{d}} = \frac{S \times (a- \epsilon)^d}{S \times a^d} =  \frac{a- \epsilon^d}{a^d} = (\frac{a- \epsilon}{a})^d = (1 - \frac{\epsilon}{a})^d $$

#### Part b:

To prove: $\forall \epsilon$, where $0<\epsilon<a$, for some a>0, $\lim_{d \to \infty}1- (1 - \frac{\epsilon}{a})^d = 1$.



###### Real analysis proof:

$\lim_{d \to \infty} 1-(1 - \frac{\epsilon}{a})^d = 1$ is logically equivalent to the statement:

$\forall \lambda >0, \exists d_1 $ such that $\forall d>d_1$, $|| 1-1+(1 - \frac{\epsilon}{a})^d|| < \lambda$, so $||(1 - \frac{\epsilon}{a})^d|| < \lambda$

-----------
##### Lemma 1:

To prove: $(1 - \frac{\epsilon}{a})^d$ is strictly decreasing.

Note, $d>d_1$, so $d = d_1 +t$ for some positive t.

So, $\frac{(1 - \frac{\epsilon}{a})^{d_1}}{(1 - \frac{\epsilon}{a})^d}$ = $\frac{(1 - \frac{\epsilon}{a})^{d_1}}{(1 - \frac{\epsilon}{a})^{d_1+t}}$ =  $\frac{(1 - \frac{\epsilon}{a})^{d_1}}{(1 - \frac{\epsilon}{a})^{d_1}(1 - \frac{\epsilon}{a})^{t}}$ = $\frac{1}{(1 - \frac{\epsilon}{a})^{t}}$ 

Since $\frac{\epsilon}{a} > 0$, it follows $1 - \frac{\epsilon}{a} <1$, so $(1 - \frac{\epsilon}{a})^t <1$ for any t, thus $\frac{1}{(1 - \frac{\epsilon}{a})^{t}} > 1$ .

So, $1>(1 - \frac{\epsilon}{a})^{t}$, thus it follows $(1 - \frac{\epsilon}{a})^{d_1}>(1 - \frac{\epsilon}{a})^{t}(1 - \frac{\epsilon}{a})^{d_1}$, making $(1 - \frac{\epsilon}{a})^{d_1}>(1 - \frac{\epsilon}{a})^{d_1 + t}$, or in other words $(1 - \frac{\epsilon}{a})^{d_1}>(1 - \frac{\epsilon}{a})^{d}$.

Thus, for every $d>d_1$, $(1 - \frac{\epsilon}{a})^{d}<(1 - \frac{\epsilon}{a})^{d_1}$.

$\therefore (1 - \frac{\epsilon}{a})^{d}$ is a strictly decreasing function.

--------

##### Lemma 2:

To prove: $(1 - \frac{\epsilon}{a})^d$ is bounded from below by 0.

Since $0<\epsilon<a$, it follows $\frac{0}{a}<\frac{\epsilon}{a}<\frac{a}{a}$, thus $0<\frac{\epsilon}{a}<1$. So $-\frac{\epsilon}{a}<\frac{\epsilon}{a}-\frac{\epsilon}{a}<1-\frac{\epsilon}{a}$. 

It follows $0 < 1 - \frac{\epsilon}{a} $, so $(1 - \frac{\epsilon}{a})^d > 0 $ for all d.

Suppose, by way of contradiction, that $(1 - \frac{\epsilon}{a})^d$ is bounded by y for $y>0$. Then $(1 - \frac{\epsilon}{a})^d \geq y > 0$ for all d. Note for $d_1=\frac{\log y}{\log (1 - \frac{\epsilon}{a})}$, $(1 - \frac{\epsilon}{a})^d = y $. Since $(1 - \frac{\epsilon}{a})^{d}$ is a strictly decreasing function, $\forall$ $d>d_1$, it follows that $(1 - \frac{\epsilon}{a})^d < y $. Therefor $(1 - \frac{\epsilon}{a})^d$ is not bounded by y, $\forall y>0$. 

Since $(1 - \frac{\epsilon}{a})^d$ is strictly decreasing, strictly positive, and not bounded by any $y>0$, it must be bounded from below by 0.

$\therefore$ $(1 - \frac{\epsilon}{a})^d$ is bounded from below by 0.

-----------------

Since $(1 - \frac{\epsilon}{a})^{d}$ is a strictly decreasing function by lemma 1, and its bounded from below by 0 by lemma 2, it follows by the Monotone Convergance Theorem that $\lim_{d \to \infty} (1 - \frac{\epsilon}{a})^d = 0$, so $1-\lim_{d \to \infty} (1 - \frac{\epsilon}{a})^d = 1- 0$.


$\therefore$ $\lim_{d \to \infty} 1 - (1 - \frac{\epsilon}{a})^d = 1$.

## 2

```python
def f_ratio(e_a, d):
    return 1-math.pow((1-e_a),d)

ea_ratio = .01
values = [2,10,1000]

print "F Ratios","\n"
for d in values:
    print "d={}: {}".format(d,f_ratio(ea_ratio,d))
```

We see that As $d$ gets larger, the $f$ ratio approaches 1.
